# Analysing Tourist Travel Pattern in Europe.
 The dataset contains information about various tourist travels across Europe. 

Here are the columns included in the dataset and what they represent:

- **Tourist_ID:** A unique identifier for each tourist.

- **Country_Visited:** The country that the tourist visited.

- **City_Visited:** The city that the tourist visited within the country.

- **Mode_of_Travel:** The mode of travel used by the tourist (e.g., Flight, Car, Train, Bicycle, Bus).

- **Travel_Duration_Days:** The total duration of the trip in days.

- **Number_of_Companions:** The number of companions the tourist traveled with.

- **Total_Travel_Cost:** The total cost of the trip in the local currency.

- **Accommodation_Type:** The type of accommodation used during the trip (e.g., Hotel, Hostel, Airbnb, Camping).

- **Main_Purpose:** The main purpose of the trip (e.g., Business, Leisure, Family Visit).

- **Season_of_Visit:** The season during which the trip took place (e.g., Winter, Spring, Summer, Fall).

## Top three countries visited

```python
# matplotlib.pyplot is used here to plot graph

import matplotlib.pyplot as plt

## Top Three Countries visited
top_countries_counts = df['Country_Visited'].value_counts().head(3)
top_countries = top_countries_counts.index.tolist()
sizes = top_countries_counts.values
print(top_countries)

# Create a pie chart
plt.figure(figsize=(8, 6))
plt.pie(sizes, labels=top_countries, colors=['green', 'blue', 'yellow'], autopct='%1.1f%%')

# Title
plt.title('Top Three Countries Visited')

# Create a pie chart
plt.show()
```

## Top cities visited

```python
## Identify the most visited cities within the top three countries

top_countries_data = df[df['Country_Visited'].isin(top_countries)]


most_visited_cities = top_countries_data.groupby(['Country_Visited', 'City_Visited']).size().reset_index(name='Visit_Count')
most_visited_cities = most_visited_cities.sort_values(by='Visit_Count', ascending=False).groupby('Country_Visited').head(1)
print(most_visited_cities)


## Plot travel duration variation among these cities

fig, axes = plt.subplots(1, 3, figsize=(19, 6), sharey=True)
for i, country in enumerate(most_visited_cities['Country_Visited']):
    city = most_visited_cities[most_visited_cities['Country_Visited'] == country]['City_Visited'].values[0]
    city_data = top_countries_data[(top_countries_data['Country_Visited'] == country) & (top_countries_data['City_Visited'] == city)]
    axes[i].hist(city_data['Travel_Duration_Days'], bins=10, color='skyblue', edgecolor='black')
    axes[i].set_title(f'Travel Duration in {city}, {country}')
    axes[i].set_xlabel('Travel Duration (Days)')
    if i == 0:
        axes[i].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

## Average total travel cost for different number of companions

```python
average_cost_per_companions = df.groupby('Number_of_Companions')['Total_Travel_Cost'].mean().reset_index()
print(average_cost_per_companions)

```

##  Plotting the average cost per number of companions

```python
# Bar chart
plt.figure(figsize=(10, 6))
plt.bar(average_cost_per_companions['Number_of_Companions'], average_cost_per_companions['Total_Travel_Cost'], color='skyblue', edgecolor='black')
plt.xlabel('Number of Companions')
plt.ylabel('Average Total Travel Cost')
plt.title('Average Total Travel Cost by Number of Companions')
plt.show()
```

```python
# Line plot

plt.figure(figsize=(10, 8))
plt.plot(average_cost_per_companions['Number_of_Companions'], average_cost_per_companions['Total_Travel_Cost'], marker='o', color='blue', linestyle='-')
plt.xlabel('Number of Companions')
plt.ylabel('Average Total Travel Cost')
plt.title('Average Total Travel Cost by Number of Companions')
plt.grid(True)
plt.show()
```

### Summary of Travel Data Analysis

#### Key Insights:

1. **Country Preferences**: Germany,Portugal and Greece emerged as the top three most visited countries. Within these countries, cities like Hamburg, Manchester, and Athens were the most frequently visited.

2. **Mode of Travel**: The most common modes of travel were flights, cars, and bicycles. Business trips predominantly used flights, while leisure trips saw a mix of bicycles and cars.

3. **Travel Duration**: The average travel duration was approximately 10.82 days. Family visits tended to have longer durations compared to business and leisure trips.

4. **Travel Companions**: There was a positive correlation between the number of companions and the total travel cost. Traveling with more people generally led to higher total costs, but the average cost per person could be lower due to shared expenses.

5. **Total Travel Cost**: The average total travel cost varied significantly based on accommodation type and mode of travel. Hotels and flights were associated with higher costs, while hostels and bicycles offered more budget-friendly options.

6. **Seasonal Trends**: The summer season had the highest number of leisure trips, while business trips were more evenly distributed across all seasons.

7. **Purpose of Travel**: The main purposes of travel were business, leisure, and family visits. The purpose of travel significantly influenced both the choice of mode of travel and the total travel cost. For example, business trips often involved flights and higher costs, while leisure trips were more budget-conscious.

8. **Accommodation Type**: Hotels were the most popular accommodation type, followed by hostels and Airbnbs. The choice of accommodation influenced the total travel cost, with hotels being the most expensive and hostels being more economical.
